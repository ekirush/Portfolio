# Join both datasets

```python
# Joining both datasets

df = pd.concat([df2,df1])
print(f"The shape after concatenating both data: {df.shape}")
df.head()
```

```python
# Rename more columns 

df = df.rename(columns={'A/W':'Attacking Work Rate',
                'D/W':'Defensive Work Rate',
                'IR':'International Reputation',
                '↓OVA':'Overall Rating ',
                'POT':'Potential'})

# Convert all column names to lowercase
df.columns = df.columns.str.lower()

df.columns
```

```python
df.head()
```

```python
# Using REGEX and replace function to remove unwanted characters from various columns

df['club'] = df['club'].str.replace('\n','').str.strip()
df['w/f'] = df['w/f'].str.replace('★','').str.strip()
df['sm'] = df['sm'].str.replace('★','').str.strip()
df['international reputation'] = df['international reputation'].str.replace('★','').str.strip()
df['value'] = df['value'].str.replace('€','').str.strip()
df['release clause'] = df['release clause'].str.replace('€','').str.strip()
df['wage'] = df['wage'].str.replace('€','').str.strip()
```

```python
# Examining the columns ['value', 'wage', 'release clause']

print(df['value'].unique())
print("-"*100)
print(df['wage'].unique())
print("-"*100)
print(df['release clause'].unique())
```

The columns ['value', 'wage', 'release clause'] contain varying SI unit symbols. These symbols need to be removed from the columns.

```python
# Removing SI units and converting 'K' to 'lbs'

# Function to convert money value to Millions
def convert_money(value):
    if 'M' in value:
        return float(value.replace('M', '')) 
    elif 'K' in value:
        return float(value.replace('K', ''))/ 1000
    return float(value)

# Columns to convert
columns_to_convert = ['value', 'wage', 'release clause']

# Apply the function 
for col in columns_to_convert:
    df[col] = df[col].apply(convert_money)
```

```python
# Examining the columns ['height', 'weight']
print(df.height.unique())
print("-"*100)
print(df.weight.unique())
```

```python
# Removing SI units and converting height to 'cm' and 'lbs' to 'kg'.

# Function to convert height
def convert_height(height_str):
    if 'cm' in height_str:
        return float(height_str.replace('cm', ''))
    elif '\'' in height_str and '\"' in height_str:
        feet, inches = height_str.split("'")
        inches = inches.strip('"')
        feet = int(feet)
        inches = int(inches)
        total_inches = feet * 12 + inches
        return total_inches * 2.54
    else:
        return None  # or raise an exception if unexpected format

# Function to convert weight
def convert_weight(weight_str):
    if 'kg' in weight_str:
        return float(weight_str.replace('kg', ''))
    elif 'lbs' in weight_str:
        pounds = float(weight_str.replace('lbs', ''))
        return pounds * 0.453592
    else:
        return None  # or raise an exception if unexpected format

# Apply conversion functions to DataFrame
df['height'] = df['height'].apply(convert_height)
df['weight'] = df['weight'].apply(convert_weight)
```

```python
# Splitting the 'contract' column into two separate columns: ['contract start', 'contract end'].

df[['contract start','contract end']] = df.contract.str.split('~',expand=True)
df = df.drop('contract',axis=1)
```

```python
# Remove '\n' on the column
df['contract start'] = df['contract start'].str.replace('\n','').str.strip()

# Extract the date and update the 'loan date end' column
df['loan date end'] = df['club'].str.extract(r'(\w+ \d{1,2}, \d{4})')

# Function to remove last 20 characters using regex
def trim_last_n_chars(s, n):
    pattern = rf'.{{{n}}}$'
    return re.sub(pattern, '', s)

# Apply the function to the 'Team_Info' column
df['club'] = df['club'].apply(lambda x: trim_last_n_chars(x, 20))

# Check the shape of ['contract start','contract end']
print(df['contract start'].unique())
print("-"*100)
print(df['contract end'].unique())
```

```python

# Remove 'On Loan' from the 'contract start' column.
df['contract start'] = df['contract start'].str.replace(r' On Loan.*', '', regex=True).str.strip()
```

```python
print(df['contract start'].unique())
print("-"*100)
print(df['contract end'].unique())
```

```python
df.head()
```

```python
# Data types converson for various columns

col1 = ['loan date end','joined','contract start','contract end'] # date

col2 = ['w/f','sm','international reputation','hits'] #int


for col in col1:
    df[col] = pd.to_datetime(df[col], errors='coerce')

for col in col2:
    df[col] = pd.to_numeric(df[col], errors='coerce').astype('Int64')
```

```python
# Checking duplicates
df.duplicated().sum()
```

```python
# Dropping the only dublicate we've in our dataset

df = df.drop_duplicates()

df.duplicated().sum()
```

```python
# Checking null values

df.isna().sum()
```

```python
# Impute contract start with the date of joining the club
df['contract start'] = df['contract start'].fillna(df['joined'])

# Impute hits with its mean
df['hits'] = df['hits'].fillna(df['hits']).mean() #loan date end	


# Impute contract_end with the date the player's loan ended
df['contract end'] = df['contract end'].fillna(df['loan date end'])
```
